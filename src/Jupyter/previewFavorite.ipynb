```python
for record in follows.find():
    following_pairs = []
    for following in record['followings']:
        pair = {}
        pair['followed'] = following['uid']
        pair['follower'] = record['fromuserid']
        following_pairs.append(pair)
    if len(following_pairs) > 0:
        result = rfollow.insert_many(following_pairs)
```
